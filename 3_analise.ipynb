```python
import pandas as pd
import matplotlib.pyplot as plt
from banco import obter_engine

engine = obter_engine()

query_1 = """
SELECT nome_orgao_superior, 
       SUM(valor_diarias + valor_passagens + valor_outros_gastos) AS custo_total
FROM silver_viagem
GROUP BY nome_orgao_superior
ORDER BY custo_total DESC
LIMIT 5;
"""

df_1 = pd.read_sql(query_1, engine)
display(df_1)

# Gerando o gráfico com os elementos obrigatórios
ax = df_1.plot(kind='bar', x='nome_orgao_superior', y='custo_total', legend=False, color='skyblue')
plt.title('Top 5 Órgãos com Maior Custo Total de Viagens')
plt.xlabel('Órgão Superior')
plt.ylabel('Custo Total (R$)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
